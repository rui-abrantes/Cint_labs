```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

def minMaxNorm(x, min, max):
    return (x - min) / (max - min)

def zScoreNorm(x, mean, sd):
    return (x- mean) / sd
```

```python
df = pd.read_csv('DCOILBRENTEUv2.csv')
df
```

```python
mean = np.mean(df['DCOILBRENTEU'])
std = np.std(df['DCOILBRENTEU'])
max = np.max(df['DCOILBRENTEU'])
min = np.min(df['DCOILBRENTEU'])
zScore = zScoreNorm(df['DCOILBRENTEU'], mean, std)
minMax = minMaxNorm(df['DCOILBRENTEU'], min, max)
```

```python



fig, (ax1, ax2) = plt.subplots(1,2)
fig.suptitle('Normalized and Standarized Prices of Oil')
fig.set_size_inches(18.5, 10.5)
fig.set_figwidth(8)
fig.set_figheight(3)


ax1.plot( minMax)

ax2.plot(zScore)
plt.show()
```

```python
#checkign moving average

minMaxMovingAverage = minMax.rolling(50).mean()

zScoreMovingAverage = zScore.rolling(50).mean()

plt.figure()
plt.plot(minMax)
plt.plot(minMaxMovingAverage)

plt.show()
```
